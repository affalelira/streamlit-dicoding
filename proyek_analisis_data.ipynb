### Cleaning Data

#### *Menangani tipe data orders*

```python
date_orders_df.info()
```

```python
datetime_columns = ['order_purchase_timestamp', 'order_approved_at','order_delivered_carrier_date',
                    'order_delivered_customer_date', 'order_estimated_delivery_date']

for column in datetime_columns:
    date_orders_df[column] = pd.to_datetime(date_orders_df[column])
```

```python
date_orders_df.info()
```

```python
date_orders_df.head()
```

**Insight:**
Pada data date_order, tipe data yang semula masih bertipe objek, dalam cleaning data diubah menjadi data yang bertipe *datetime* agar dapat dilakukan proses analisis selanjutnya 

#### *Menangani missing value pada data orders*

```python
date_orders_df.isna().sum()
```

```python
missing_values = date_orders_df[date_orders_df.isna().any(axis=1)]
missing_values.head(10)
```

**Insight:**
Terlihat pada hasil missing value diatas, nilai yang kosong berada pada order status 'invoice', 'shipped', 'processing' dan 'unavailable'. Dalam project kali ini, hanya akan mengambil order status 'delivery' sebagai proses analisis berikutnya sehingga nilai missing value tersebut akan di drop. 

```python
status_drop = date_orders_df[date_orders_df['order_status'].isin(['invoiced', 'shipped', 'processing', 'unavailable','canceled','created','approved'])].index

orders_df_delivered = date_orders_df.drop(status_drop)
```

```python
orders_df_delivered.isna().sum()
```

```python
missing_values1= orders_df_delivered[orders_df_delivered.isna().any(axis=1)]
missing_values1.head(10)
```

**Insight:**
Terlihat pada hasil drop status order, masih terdapat tanggal order delivered yang bernilai missing value, dalam hal ini akan diisi dengan nilai mean untuk setiap missing value

```python
mean_date_column = ['order_approved_at','order_delivered_carrier_date','order_delivered_customer_date']

for column in mean_date_column:
    mean_date = orders_df_delivered[column].mean()  
    orders_df_delivered[column].fillna(mean_date, inplace=True)

orders_df_delivered.head(10)
```

```python
orders_df_delivered.isna().sum()
```

**Insight:**
Nilai missing value telah tergantikan oleh nilai mean dari setiap kolom. Nilai tersebut akan digunakan untuk proses analisis data selanjutnya

## Analisis Lanjutan RFM Analysis

```python
merged_df = pd.merge(
    orders_df_delivered,                
    payment_orders_products_df,        
    on='order_id',                      
    how='inner'                         
)

merged_df.head()
```

```python
final_merged_df = pd.merge(
    merged_df,                      
    customers_df,                   
    on='customer_id',               
    how='inner'                    
)

final_merged_df.head()
```

```python
final_merged_df.info()
```

```python
rfm_df = final_merged_df.groupby(by="customer_unique_id", as_index=False).agg({
    "order_purchase_timestamp": "max",  
    "order_id": "nunique",             
    "payment_value": "sum"              
})

rfm_df.columns = ["customer_unique_id", "max_order_timestamp", "frequency", "monetary"]

rfm_df["max_order_timestamp"] = rfm_df["max_order_timestamp"].dt.date
recent_date = merged_df["order_purchase_timestamp"].dt.date.max()  
rfm_df["recency"] = rfm_df["max_order_timestamp"].apply(lambda x: (recent_date - x).days)

rfm_df.drop("max_order_timestamp", axis=1, inplace=True)

rfm_df.head()
```

```python
# Memperpendek customer_unique_id untuk tampilan yang lebih baik
rfm_df['short_customer_id'] = rfm_df['customer_unique_id'].apply(lambda x: f"{x[:5]}...{x[-5:]}")

fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(30, 6))

colors = ["#72BCD4"] * 5

sns.barplot(y="frequency", x="short_customer_id", data=rfm_df.sort_values(by="frequency", ascending=False).head(5), palette=colors, ax=ax[0])
ax[0].set_ylabel(None)
ax[0].set_xlabel(None)
ax[0].set_title("By Frequency", loc="center", fontsize=18)
ax[0].tick_params(axis='x', labelsize=15)
ax[0].tick_params(axis='x', rotation=45)  

sns.barplot(y="monetary", x="short_customer_id", data=rfm_df.sort_values(by="monetary", ascending=False).head(5), palette=colors, ax=ax[1])
ax[1].set_ylabel(None)
ax[1].set_xlabel(None)
ax[1].set_title("By Monetary", loc="center", fontsize=18)
ax[1].tick_params(axis='x', labelsize=15)
ax[1].tick_params(axis='x', rotation=45)  

plt.suptitle("Best Customer Based on RFM Parameters (customer_id)", fontsize=20)
plt.tight_layout()
plt.show()
```

**Insight**
1. **Frekuensi (Frequency)**
   - **Customer ID** `8d60f...c8455` adalah pelanggan yang paling sering melakukan transaksi, dengan jumlah frekuensi tertinggi.
   - **Customer ID** `3e43e...cf44c`, `1b67c...9553`, `ca770...46268`, dan `6469f...f1782` memiliki frekuensi yang lebih rendah dibandingkan dengan pelanggan yang teratas. Ini menunjukkan bahwa mereka tidak melakukan transaksi sebanyak pelanggan dengan ID `8d60f...c8455`.
   - Dari visualisasi ini dapat menunjukkan bahwa pelanggan yang lebih aktif memiliki frekuensi transaksi yang lebih tinggi dimana menunjukkan loyalitas atau kepuasan yang lebih besar terhadap produk atau layanan.

2. **Monetary**
   - Pada monetary, **Customer ID** `0a09...aa872` menghasilkan nilai monetary tertinggi. Artinya setiap pembelian atau transaksi yang dilakukan (walaupun tidak sering) berjumlah harga yang sangat tinggi. 
   - Diikuti oleh **Customer ID** `698e1...ad6f8` dan `c402f...44fb`, yang juga memiliki nilai monetary yang cukup tinggi.
   - Pelanggan dengan nilai monetary yang tinggi berkontribusi lebih signifikan terhadap pendapatan perusahaan, meskipun frekuensi transaksi mereka mungkin tidak setinggi pelanggan lainnya.

**Recency**

Selanjutnya analisis recency akan dibedakan dan dilakukan untuk mengidentifikasi distribusi waktu sejak transaksi terakhir setiap pelanggan. Mengingat volume data yang besar, visualisasi dalam bentuk histogram digunakan untuk mempermudah interpretasi.
